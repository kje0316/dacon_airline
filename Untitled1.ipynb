```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
sample_df = pd.read_csv('sample_submission.csv')
```

```python
train_df.drop(['Cancelled','Diverted'], axis=1, inplace= True) #Cancelled와 Diverted 데이터는 모든 데이터가 0으로 존재 --> drop 
```

```python
train_df = train_df[train_df['Estimated_Arrival_Time']>train_df['Estimated_Departure_Time']] #도착시간이 출발시간보다 큰 경우만tra
```

```python
def find_time_lag(train_df, state):
    OA = train_df[train_df['Origin_State'] == state]
    predict_st = OA.Destination_State.unique().tolist()
    DA = train_df[train_df['Destination_State'] == state]
    time_delay_dict = {}
    for st in predict_st:
        st_oa = OA[OA['Destination_State'] == st]
        st_oa_diff = st_oa.Estimated_Arrival_Time.mean() - st_oa.Estimated_Departure_Time.mean()
        st_da = DA[DA['Origin_State'] == st]
        st_da_diff = st_da.Estimated_Arrival_Time.mean() - st_da.Estimated_Departure_Time.mean()
        time_delay = st_da_diff - st_oa_diff
        if not np.isnan(time_delay):
            time_delay_dict[state, st] = -round(time_delay/2/60)
    
    return time_delay_dict

time_delay = dict()
for i in train_df['Origin_State'].unique().tolist():
    time_delay.update(find_time_lag(train_df, i))
    
print(time_delay)
```

```python
train_df['time_diff'] = np.nan
```

```python
# 구한 시차를 df에 time_diff 컬럼에 넣어주기 
for i in range(len(train_df)):   
    if (train_df['Origin_State'].iloc[i],train_df['Destination_State'].iloc[i]) in time_delay.keys() :
        train_df['time_diff'].iloc[i] = time_delay[train_df['Origin_State'].iloc[i],train_df['Destination_State'].iloc[i]]
    else : 
        train_df['time_diff'].iloc[i] == 'time'
```

```python
train_df['flight_time'] = (train_df['Estimated_Arrival_Time']-train_df['Estimated_Departure_Time'])#-
```

```python
train_df['flight_time'] = (train_df['flight_time']//100)*60 + train_df['flight_time']%100   #도착시간-출발시간 분으로 표시
```

```python
train_df['flight_time'] = train_df['flight_time'] - train_df['time_diff']*60  #시차계산 포함
```
